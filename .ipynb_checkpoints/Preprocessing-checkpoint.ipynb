```python
def extract_year(date):
    idx = date.find(', ')
    year = int(date[idx+2:])
    return year
players_cv['Birth Date'].apply(lambda d: extract_year(d))
```

```python
import re

def preprocess_name(name):
    if ',' in name:
        ind = name.find(',')
        name = name[:ind]
    ### We remove the Jr.
    if "Jr." in name:
        name = name.replace("Jr.", "")
    if "Jr" in name:
        name = name.replace("Jr.", "")
    if "III" in name:
        name = name.replace("III", "")
    while '.' in name:
        ind = name.find('.')
        name = name.replace('.','')
    name = name.lower()
    
    return name  
```

```python
import warnings
warnings.filterwarnings('ignore')
salary_15_16.loc[:, 'NAME'] = salary_15_16['NAME'].apply(lambda s: preprocess_name(s))
```
